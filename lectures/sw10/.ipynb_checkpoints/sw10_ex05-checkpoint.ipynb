```python
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
def calculate_points(n, R_a, R_i):
    x = []
    y = []
    z = []
    r = R_a - R_i
    for u_ in range(n):
        u = (2 * np.pi) / n * u_
        for v_ in range(n):
            v =  (2 * np.pi) / n * v_

            x_u = (R_i + R_a) / 2 * np.array([np.cos(u), np.sin(u), 0]).T

            
            n_u =  np.array([-np.cos(u), -np.sin(u), 0]).T
            b_u =  np.array([0, 0, 1]).T

            s_u = r * (n_u * np.cos(v) + b_u * np.sin(v))
            S_u = x_u + s_u
            
            x.append(S_u[0])
            y.append(S_u[1])
            z.append(S_u[2])
            
    return (x,y,z)
```

```python
def plot_3D(x,y,z):
    fig = plt.figure(figsize=(4,4))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x,y,z) # plot the point on the figure
    ax.set_xlim(-15,15)

    ax.set_ylim(-15,15)

    ax.set_zlim(-15,15)


    plt.show()
```

```python
x,y,z = calculate_points(16, 10, 14)
plot_3D(x,y,z)
```
