## Geraden und ihre Schnittpunkte

### Punkt-Richtungsform einer Geraden

#### mathematisch:

##### vektoriell

$
\mathbf{r}
    = \mathbf{p} + s \cdot \mathbf{u}
    = \mathbf{p} + s \cdot \left[ \mathbf{q} - \mathbf {p} \right]
$

##### komponentenweise

$
\begin{pmatrix} x \\ y \end{pmatrix}
    = \begin{pmatrix} p_1 \\ p_2 \end{pmatrix}
        + s \cdot \begin{pmatrix} u_1 \\ u_2 \end{pmatrix}
    = \begin{pmatrix} p_1 \\ p_2 \end{pmatrix}
        + s \cdot \left[ \begin{pmatrix} q_1 \\ q_2 \end{pmatrix} 
            - \begin{pmatrix} p_1 \\ p_2 \end{pmatrix} \right]
    = \begin{pmatrix} p_1 \\ p_2 \end{pmatrix}
        + s \cdot \begin{pmatrix} q_1 - p_1 \\ q_2 - p_2 \end{pmatrix} 
$

#### und in Python:

```python
def Line (p, q, t):
    return p + t * (q - p)
```

```python
def Points (*p):
    q = Cat (*p)
    pl.plot (q [0], q [1], 'o')
```

```python
p1 = Line (a, b, 0.5)
#p2 = Line (b, c, 0.25)
#p3 = Line (c, a, 0.75)
#p4 = Line (c, a, 0.1)
#p5 = Line (a, b, 2)
#p6 = Line (a, b, -1)
Curve (a, b, c, a)
#Points (a, b, c, p1, p2, p3, p4, p5, p6)
Points (a, b, c, p1)
Iso()
```

### Schnittpunkt zweier Geraden

#### Werkzeug: 2x2-Gleichungssystem lösen (Cramersche Regel)

$
\begin{align}
    ax + by 
        &= c 
        & \cdot d \\
    dx + ey 
        &= f 
        & \cdot a \\[6pt]
    adx + bdy 
        &= cd 
        & \mathrm{(I)} \\
    adx + aey 
        &= af 
        & \mathrm{(II)} \\
    aey - bdy 
        &= af - cd 
        & \mathrm{(II - I)} \\[6pt]
    y 
        &= \displaystyle \frac{af - cd}{ae - bd} 
        = \displaystyle \frac{\begin{vmatrix} a & c \\ d & f \end{vmatrix}}{
            \begin{vmatrix} a & b \\ d & e \end{vmatrix}} \\[6pt]
    x 
        &= \displaystyle \frac{\begin{vmatrix} c & b \\ f & e \end{vmatrix}}{
            \begin{vmatrix} a & b \\ d & e \end{vmatrix}} 
        = \displaystyle \frac{cf - be}{ae - bd} 
        & \qquad \mathrm{(analog)}
    \\[6pt]
\end{align}
$

#### Zwei Geraden:

$
\begin{align}
\mathbf{r} &= \mathbf{p} + s \cdot \mathbf{u} & \qquad \mathrm{1.~Gerade} \\
\mathbf{r} &= \mathbf{q} + t \cdot \mathbf{v} & \qquad \mathrm{2.~Gerade}\\
\end{align}
$

#### Schnittpunkt der Geraden

#####  mathematisch:

$
\begin{align}
\mathbf{p} + s \cdot \mathbf{u} &= \mathbf{q} + t \cdot \mathbf{v} \\
s \cdot \mathbf{u} - t \cdot \mathbf{v} &= \mathbf{q} - \mathbf{p} \\
\end{align}
$

##### in Python:

```python
def Det2 (a, b):
    return la.det (Cat (a, b))
```

```python
def LineSect (p1, q1, p2, q2):
    u1 = q1 - p1
    u2 = p2 - q2
    d = Det2 (u1, u2)
    d1 = Det2 (p2 - p1, u2)
    s1 = d1 / d
    return s1
```

```python
p = Vec (3, 5)
q = Vec (4, 1)
s1 = LineSect (p, q, a, b); p1 = Line (p, q, s1)
s2 = LineSect (p, q, b, c); p2 = Line (p, q, s2)
s3 = LineSect (p, q, c, a); p3 = Line (p, q, s3)
Curve (a, b, c, a)
Curve (p, q)
Points (p1, p2, p3)
Iso()
```

### Schnittpunkt eines Segmentes mit einer Geraden

```python
def SegSect (p1, q1, p2, q2):
    u1 = q1 - p1
    u2 = p2 - q2
    w = p2 - p1
    d = Det2 (u1, u2)
    if d == 0:
        return None
    s1 = Det2 (w, u2) / d
    if s1 < 0 or s1 > 1:
        return None
    s2 = Det2 (u1, w) / d
    if s2 < 0 or s2 > 1:
        return None
    return s1
```

```python
s = { None }
s.add (SegSect (p, q, a, b))
s.add (SegSect (p, q, b, c))
s.add (SegSect (p, q, c, a))
s = list (filter (None, s))
s.sort()

Curve (a, b, c, a)
Curve (p, q)
Points (Line (p, q, s))
Iso()
```

## Darstellung einer Fläche als Drahtgittermodell

### Katenoid

```python
def F (u, v):
    r = np.cosh (v)
    x = r * np.cos (u)
    y = r * np.sin (u)
    f = np.stack ((x, y, v), axis = 0)
    return f
```

### Grundriss

```python
u = np.linspace (0, 2 * np.pi, 50)
v = np.linspace (-1.5, 1.5, 25)
uu, vv = np.meshgrid (u, v)
pp = F (uu, vv)
```

```python
def Curve (*p):
    q = Cat (*p)
    pl.plot (q [0], q [1], color = 'black', linewidth = 0.3)
```

```python
def Wireframe (p):
    for i in range (p.shape [1]):
        Curve (p [:, i, :])
    for k in range (p.shape [2]):
        Curve (p [:, :, k])
```

```python
xy = np.stack ((pp [0], pp [1]), axis = 0)
Wireframe (xy)
Iso()
```

### Seitenriss

```python
xz = np.stack ((pp [0], pp [2]), axis = 0)
Wireframe (xz)
Iso()
```

### Schrägbild

#### Rotationen definieren

##### mathematisch:

$
\begin{align}
    R_x (\varphi)
        &= \begin{pmatrix} 
            1 & 0 & 0 \\ 
            0 & \cos\varphi & -\sin\varphi \\ 
            0 & \sin\varphi & \cos\varphi \end{pmatrix}
        & \mbox{(Rotation um die $x$-Achse)}
        \\[12pt]
    R_y (\varphi)
        &= \begin{pmatrix} 
            \cos\varphi & 0 & \sin\varphi \\ 
            0 & 1 & 0 \\ 
            -\sin\varphi & 0 & \cos\varphi \end{pmatrix}
        & \mbox{(Rotation um die $y$-Achse)}
        \\[12pt]
    R_z (\varphi)
        &= \begin{pmatrix} 
            \cos\varphi & \sin\varphi & 0 \\ 
            -\sin\varphi & \cos\varphi & 0 \\
            0 & 0 & 1 \end{pmatrix}
        & \mbox{(Rotation um die $z$-Achse)}
\end{align}
$

##### in Python:

```python
def RotX (phi):
    t = phi * np.pi / 180
    c = np.cos (t)
    s = np.sin (t)
    return np.array ([[1, 0, 0], [0, c, -s], [0, s, c]])
```

```python
def RotY (phi):
    t = phi * np.pi / 180
    c = np.cos (t)
    s = np.sin (t)
    return np.array ([[c, 0, s], [0, 1, 0], [-s, 0, c]])
```

```python
def RotZ (phi):
    t = phi * np.pi / 180
    c = np.cos (t)
    s = np.sin (t)
    return np.array ([[c, -s, 0], [s, c, 0], [0, 0, 1]])
```

```python
A = RotZ (30) @ RotX (40)
qq = np.tensordot (A, pp, axes = 1)
Wireframe (qq)
Iso()
```

## Schattierte Darstellung einer Fläche 

### Mathematische Werkzeuge

#### Skalarprodukt

```python
def Dot (v, w):
    return v [0, 0] * w [0, 0] + v [1, 0] * w [1, 0] + v [2, 0] * w [2, 0]
```

#### Länge (Norm) eines Vektors

```python
def Norm (v):
    return np.sqrt (Dot (v, v))
```

#### Vektor normalisieren

```python
def Normalized (v):
    n = Norm (v)
    if n == 0:
        return v
    return v / n
```

#### Vektor- oder Kreuzprodukt

```python
def Cross (v, w):
    x = v [1, 0] * w [2, 0] - v [2, 0] * w [1, 0]
    y = v [2, 0] * w [0, 0] - v [0, 0] * w [2, 0]
    z = v [0, 0] * w [1, 0] - v [1, 0] * w [0, 0]
    return Vec (x, y, z)
```

### Dreieck schattiert darstellen

```python
def TriangleShaded (a, b, c, light):
    u = b - a
    v = c - a
    n = Normalized (Cross (u, v))
    col = abs (Dot (light, n))
    p = Cat (a, b, c)
    pl.fill (p [0], p [1], color = (col, col, col))
```

### Fläche schattiert darstellen (Algorithmus des Malers)

```python
def FlatShaded (pp, light):
    light = Normalized (light)
    for i0 in range (pp.shape [1] - 1):
        i1 = i0 + 1
        for k0 in range (pp.shape [2] - 1):
            k1 = k0 + 1
            a = pp [:, i0, k0] .reshape (-1, 1)
            b = pp [:, i1, k0] .reshape (-1, 1)
            c = pp [:, i1, k1] .reshape (-1, 1)
            d = pp [:, i0, k1] .reshape (-1, 1)
            TriangleShaded (a, b, c, light)
            TriangleShaded (a, c, d, light)
```

```python
FlatShaded (qq, Vec (-1, 1, 1))
Iso()
```
